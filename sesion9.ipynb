FUNCION QUE VERIFICAR QUE UN STR ES O NO UN PALINDROMO

```python
l="taco cat"
l1=list(l.replace(" ","").lower())
g=list(reversed(l.replace(" ","").lower()))
print(list(l),l1,g)
if l1==g:
  print("es un palindromo")
else:
  print("no es un palindromo")
#l[-1:len(l)-1:-1] #lista[inicio:fin-1:paso]
```

```python
def palindromo(cadena):
  n_cadena=cadena.replace(" ","").lower()
  cadena2=""
  for i in range(len(n_cadena)):
    cadena2=cadena2+n_cadena[-i-1]
  return cadena2==n_cadena
```

Funcion que cuente cuantas Mayus y minus hay en un STR

```python
intro_nombre="Hola mundo mi nombre es John"
def mayus_minus(cadena):
  mayus=0
  minus=0
  for i in cadena:
    if i.isalpha():
      if i.islower():
        minus=minus+1
      else:
        mayus=mayus+1
  return mayus,minus
```

COMO SE CREA Un GRAFICO EN MATPLOTLIB:

*   Crear el lienzo
*   Definir la graficas que van a parecer en un mismo lienzo
*   (opcional) Darle caracteristicas personalizadas   
*   Visualizar




```python
import numpy as np
```

```python
x=np.arange(1,10+1)
y=x**2
```

```python
#crear el lienzo 
plt.figure()
#definir los graficos
plt.plot(x,y)
#caracteristicas adicionales 

#visualizar
plt.show()
```

DOS GRAFICAS

```python
y1=25*np.exp(-x**2)
#crear el lienzo 
plt.figure()
#definir los graficos
plt.plot(x,y)
plt.plot(x,y1)
#caracteristicas adicionales 

#visualizar
plt.show()
```

CONFIGURACIONES MINIMAS 

```python
#crear el lienzo 
plt.figure()
#definir los graficos
plt.plot(x,y,label="$x^2$",color="chartreuse",alpha=0.5) #color=(R,G,B) #alpha(transparencia)
plt.plot(x,y1,label="$25\exp(x^2)$",color="magenta")
#caracteristicas adicionales 
plt.grid()
plt.legend(loc="best")
#nombres de los ejes
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
#visualizar
plt.show()
```

MARCADORES 

```python
#crear el lienzo 
plt.figure()
#definir los graficos
plt.plot(x,y,"-",label="$x^2$",color="chartreuse",alpha=0.5) #"--"(marcador)"--b"(marcador color)
plt.plot(x,y1,marker="$x$",label="$25\exp(x^2)$",color="magenta") #$latex$ maracador en latex
#caracteristicas adicionales 
plt.grid()
plt.legend(loc="best")
#nombres de los ejes
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
#visualizar
plt.show()
```

GRAFICAS DE AREA

```python
plt.figure() #creacion del lienzo
plt.fill_between(x,y,y1,color="r",alpha=0.5) #parametros adicionales 
plt.plot(x,y,label="$x^2$",color="g")
plt.plot(x,y1,label="$25\exp(x^2)$",color="b")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="best")
plt.show() #mostrar la grafica
```

USO DEL PARAMETRO where= (filtro)

```python
x=np.linspace(0,10,100)
y=x**2
y1=25*np.exp(-x**2)
filtro= x>1.6
plt.figure() 
plt.fill_between(x,y,y1,color="k",alpha=0.5,where=filtro)  
plt.plot(x,y,label="$x^2$",color="g")
plt.plot(x,y1,label="$25\exp(x^2)$",color="b")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="best")
plt.show()
```

WHERE (dos filtros)

```python
x=np.linspace(0,10,100)
y=x**2
y1=25*np.exp(-x**2)
filtro= x>1.6
plt.figure() 
plt.fill_between(x,y,y1,color="k",alpha=0.7,where=filtro) 
plt.fill_between(x,y,y1,color="r",alpha=0.7,where=~filtro) 
plt.plot(x,y,label="$x^2$",color="g")
plt.plot(x,y1,label="$25\exp(x^2)$",color="b")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="best")
plt.show()
```

TRES FILTROS

```python
x=np.linspace(0,10,100)
y=x**2
y1=25*np.exp(-x**2)
filtro= (x>=0) & (x<=2.5)
filtro1= (x>2.5) & (x<=6.5)
filtro2= x>6.5 
plt.figure() 

plt.fill_between(x,y,y1,color="yellow",alpha=0.7,where=filtro) 
plt.fill_between(x,y,y1,color="b",alpha=0.7,where=filtro1)
plt.fill_between(x,y,y1,color="r",alpha=0.7,where=filtro2) 
plt.plot(x,y,label="$x^2$",color="k")
plt.plot(x,y1,label="$25\exp(x^2)$",color="k")

plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="best")

plt.show()
```

TEXTO EN UNA GRAFICA

```python
x=np.linspace(0,10,100)
y=x**2
y1=25*np.exp(-x**2)
filtro= (x>=0) & (x<=2.5)
filtro1= (x>2.5) & (x<=6.5)
filtro2= x>6.5 
plt.figure()

plt.fill_between(x,y,y1,color="yellow",alpha=0.7,where=filtro) 
plt.fill_between(x,y,y1,color="b",alpha=0.7,where=filtro1)
plt.fill_between(x,y,y1,color="r",alpha=0.7,where=filtro2) 
plt.plot(x,y,label="$x^2$",color="k")
plt.plot(x,y1,label="$25\exp(x^2)$",color="k")

plt.text(1.6,30,"Intercepto",ha="center",fontsize=10) #ha (ajuste horizontal ) va (ajuste verticaL) fontsize (tamaño de funte)
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="best")

plt.show()
```

PRINT ARROW

```python
plt.figure()

plt.fill_between(x,y,y1,color="yellow",alpha=0.7,where=filtro) 
plt.fill_between(x,y,y1,color="b",alpha=0.7,where=filtro1)
plt.fill_between(x,y,y1,color="r",alpha=0.7,where=filtro2) 
plt.plot(x,y,label="$x^2$",color="k")
plt.plot(x,y1,label="$25\exp(x^2)$",color="k")

plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="best")
plt.gca().annotate(   #crear una flecha
    "Intercepto",
    xy=(1.6,5),
    xytext=(2,50),
    va="center",
    ha="center",
    arrowprops=dict(facecolor="k",shrink=0.005)
)    

plt.show()
```

BARRAS MULTIPLES

```python
def poner_etiqueta(edad):
  if 0<edad<12:
    return "< 12"
  elif 12<=edad<18:
    return "[12,18)"
  elif 18<=edad<40:
    return "[18,40)"
  elif edad>=40:
    return "viejos"
  else:
    return "Error de Edad"      
```

```python
edad_etiqueta=titanic.Age.apply(poner_etiqueta)
titanic["edad_etiqueta"]=edad_etiqueta 
titanic.head() 
```

```python
titanic.groupby(["edad_etiqueta","Pclass"]).Sex.value_counts().unstack(1).plot.bar()
```

```python
titanic.groupby(["edad_etiqueta"]).Sex.value_counts().unstack(1).plot.bar()
```

```python
etiquetas=[]
for ind,val in zip(conteo_clases.index,conteo_clases.values):
  etiquetas.append("Clase"+ str(ind)+"\n" +str(val))
etiquetas
```
